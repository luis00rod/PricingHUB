## Analysis Function 

The `analysis()` function takes a GroupBy object representing a customer segment and returns a DataFrame with summary statistics about that segment.

The function does the following:

1. Creates an empty DataFrame to hold the summary stats 
2. Extracts the minimum, maximum, 10th and 90th quantile price points for the segment using `.X_price.min()`, `.X_price.max()`, `.X_price.quantile(q=0.1)` and `.X_price.quantile(q=0.9)` respectively. 
3. Sums the total `delta_price` (differences between consecutive price points) and records the result using `.X_price.sum()`
4. Sorts the prices in the segment in ascending order using `.X_price.sort_values()` 
5. Calculates additional metrics like:
    - Mean price using `.X_price.mean()`
    - Cluster size (number of data points) using `.shape[0]`
    - Total profit using `.profit.sum()`
    - Total traffic using `.traffic.sum()`
6. Rounds each metric to 3-4 decimal places for readability using `round()` 
7. Returns the summary statistics DataFrame

```python
def analysis(group):
    """
    Args:
        group (DataFrame GroupBy object): GroupBy object representing a segment.

    Returns:
        DataFrame: Summary statistics for the segment.
    """
    data = pd.DataFrame([{
                        "min_price": round(group.X_price.min(), 4), 
                        "max_price": round(group.X_price.max(), 4),
                        "quantile_10": round(group.X_price.quantile(q=0.1), 3),
                        "quantile_90": round(group.X_price.quantile(q=0.9), 3), 
                        "sum_delta_price": round(group.X_price.sum(), 3), 
                        "delta_price": group.X_price.sort_values(),  # Pandas series of all X_price sorted given an natural
                                                                    #  segment by weekday and channel
                        "mean_price": round(group.X_price.mean(), 4),
                        "cluster size": group.shape[0],
                        "sum_profit": group.profit.sum(),
                        "traffic": group.traffic.sum()}])
    return data
```

## Clustering Function

### Overview 

The `clustering()` function takes a raw DataFrame and performs K-Means clustering to assign each data point to 1 of 4 clusters. It returns the input DataFrame with an added `cluster` column indicating the cluster assignment for each row.

The function performs the following steps:

1. Selects the features to use for clustering:  `'weekday', 'X_price', 'pricing_channel', 'traffic', 'orders'`

2. Extracts only those features into `X`

3. Encodes the categorical `'weekday'` and `'pricing_channel'` features using `pd.get_dummies()`

4. Initializes a KMeans model with 4 clusters using `KMeans(n_clusters=4, random_state=42)`

5. Fits the model on `X_encoded`, the encoded DataFrame

6. Adds the cluster assignments from the model as a new column called `'cluster'` in the raw input data 

7. Returns the final clustered DataFrame

```python
def clustering(raw_data):
    """
    Args:
        raw_data (DataFrame): Input DataFrame containing the raw data.

    Returns:
        DataFrame: DataFrame with an additional 'cluster' column indicating the assigned cluster for each data point.
    """
    features = ['weekday', 'X_price', 'pricing_channel', 'traffic', 'orders']
    X = raw_data[features]
    X_encoded = pd.get_dummies(X, columns=['weekday', 'pricing_channel'])
    kmeans = KMeans(n_clusters=4, random_state=42)
    kmeans.fit(X_encoded)
    raw_data['cluster'] = kmeans.labels_
    return raw_data
```

## Plotter Function

### Overview

The `plotter()` function takes a DataFrame and creates a dual-axis line plot showing both `profit` and `X_price` over time. It employs several techniques to improve the visualization and add clarity.

The function performs the following steps:

1. Defines a figure and Axes using `plt.subplots()`
2. Creates a twin Axes to plot a second y-axis using `ax2 = ax.twinx()`
3. Plots `profit` over `date` on the first Axes `ax` 
4. Plots `X_price` over `date` on the second Axes `ax2`
5. Rotates the x-tick labels 45 degrees with `plt.xticks(rotation=45)` to prevent overlap
6. Increases the padding at the bottom of the plot using `plt.gcf().subplots_adjust(bottom=0.2)` to make room for the rotated labels
7. Moves the legend outside of the plot area using `leg = ax.legend(loc='upper left')` for better visibility  
8. Offsets the `Profit` y-axis label slightly using `ax.set_ylabel('Profit', rotation=90, labelpad=20)` to avoid complete overlap with the `X_price` label
9. Sets the remaining Axes labels 
10. Shows the final plot

### Interpretation

This function creates a clean dual-axis line plot that clearly shows how both `profit` and `X_price` are changing over time. Some of the key insights this visualization provides include:

- How are `profit` and `X_price` related? Do they follow a similar or different pattern? This could indicate how sensitive customers are to price changes. 
- Are there any time periods where `profit` drops off significantly even though `X_price` remains steady or increases? This could highlight external factors impacting customer demand.  
- How volatile is `X_price` over time? Does it change frequently or remain fairly stable? This provides context on the company's pricing strategy and approach.  

```python
def plotter(df):
    fig, ax = plt.subplots()  
    ax2 = ax.twinx()   
    
    df.plot(x='date', y='profit', kind='line', color='b', figsize=(8, 5),  
            label='Profit', ax=ax)
    df.plot(x='date', y='X_price', kind='line', color='orange',  
            label='X_price', ax=ax2)  
            
    plt.xticks(rotation=45)            # Rotate x-tick labels 
    plt.gcf().subplots_adjust(bottom=0.2)  # Increase bottom padding
    leg = ax.legend(loc='upper left')    # Move legend outside plot
    leg.get_frame().set_edgecolor('k') 
    
    ax.set_ylabel('Profit', rotation=90, labelpad=20) # Offset Profit label 
    ax2.set_ylabel('X_price') 
    
```

# Linear Response Plot Function

### Overview

The `linear_response_plot()` function performs a linear regression analysis to model the relationship between price and response (orders) for a given dataset. It plots the data points along with the resulting linear regression line, providing a visualization of the demand curve and price elasticity.

This type of price response analysis provides insights into:

- How sensitive demand is to price changes for the overall market or a particular segment
- Whether demand is elastic or inelastic based on the magnitude of the slope coefficient 
- The validity of the linear model in capturing the true relationship  (Curved models may be needed for some datasets)

```python
def linear_response_plot(raw_data_clusterized):

    # Create a new column for the response variable (e.g., orders)
    raw_data_clusterized['response'] = raw_data_clusterized['orders']

    # Create a column for the price variable (X_price)
    raw_data_clusterized['price'] = raw_data_clusterized['X_price']

    # Fit a linear regression model
    X = raw_data_clusterized[['price']]
    y = raw_data_clusterized['response']
    model = sm.OLS(y, sm.add_constant(X)).fit()

    # Extract the coefficients
    intercept = model.params[0]
    slope = model.params[1]

    x = np.linspace(raw_data_clusterized['price'].min(), raw_data_clusterized['price'].max(), 100)
    y = intercept + slope * x

    # Plot the data points
    plt.scatter(raw_data_clusterized['price'], raw_data_clusterized['response'], label='Data')

    # Plot the linear price response function
    plt.plot(x, y, color='red', label='Linear Price Response')

    # Add labels and title
    plt.xlabel('Price')
    plt.ylabel('Response')
    plt.title('Linear Price Response Function')

    # Add legend
    plt.legend()

    # Show the plot
    plt.show()
```

# Price Tracker Function 

### Overview

The `price_tracker()` function takes a row of data and calculates the current price based on the `X_price` and an initial reference price. It updates the price history and initial price for the next iteration. This function enables modeling how pricing evolves over time.

The function executes the following steps:

1. Uses the global `initial_price` variable as the starting reference price 
2. Accesses the `X_price` for the current row which represents the percentage change from the initial price
3. Calculates the `current_price` by applying `X_price` to the `initial_price` and adding it 
4. Appends the `current_price` to the global `price_history` list to track how price is evolving over time 
5. Updates `initial_price` to the current `current_price` calculated so it can be used as the reference for the next iteration.
6. Returns the input `row` with the added `current_price` column 

### Use Case

This function enables modeling an artificial price trajectory over time to gain insights into:

- How strategic choices like frequent sales or price hikes may impact the long term profitability of different products.  

```python
def price_tracker(row):
    """
    Calculate the current price based on the X_price and update the price history.

    Args:
        row (Series): Input row containing 'X_price', 'profit', and other columns.

    Returns:
        Series: The input row with the current price calculated and added.
    """
    global initial_price
    global price_history

    current_price = row['X_price'] * initial_price + initial_price  # Calculate the current price
    price_history.append(current_price)  # Add the current price to the price history list
    initial_price = current_price  # Update the initial price for the next iteration
    return row
```
